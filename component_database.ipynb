```python
comp_array = [[]]*len(components)
frac_array = [[]]*len(components)
index = 0
for comp in components:
    comp_array[index] = comp[0]
    frac_array[index] = comp[1]
    index = index +1

comp_array = np.array(comp_array, dtype=object).T
frac_array = np.array(frac_array, dtype=object).T
print(comp_array)
print(frac_array)
```

```python
comp_data = pd.DataFrame.from_records([comp_array, frac_array], nrows=len(components), index=["Stroke Point List", "Fractional Distances"])
#print(frac_array)
comp_data.head()
```
